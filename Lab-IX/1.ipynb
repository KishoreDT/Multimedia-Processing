```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
import skimage.color as cl
img=io.imread("img.png")
img_gr=cl.rgb2gray(img)
img_fft=np.fft.fftshift(np.fft.fft2(img_gr))
imgfft=np.fft.fftshift(np.fft.fft2(img_gr))
x,y=img_fft.shape
print(x,y)
m,n=int(x/2),int(y/2)
img_fft[m-1:m+1,n-1:n+1]=imgfft[0:m-6,0:y]=imgfft[m+5:x-1,0:y-1]=imgfft[m-5:m+5,0:n-6]=imgfft[m-5:m+5,n+6:y-1]=0
img_recons=abs(np.fft.ifft2(img_fft))
img_recons1=abs(np.fft.ifft2(imgfft))
t=["Orginal Image","Grayscaled Image","Magnituded Spectrum 1","Magnituded Spectrum 2","Reconstructed Image 1","Reconstructed Image 2"]
im=[img,img_gr,np.log(abs(img_fft)),np.log(abs(imgfft)),img_recons,img_recons1]
for i in range(1):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
for i in range(1,6):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```
