```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
img = io.imread("1.png")
from skimage.transform import downscale_local_mean as dlm
factors = 3**np.arange(1, 5)
figure, axis = plt.subplots(1, len(factors), figsize=(20, 6))
for factor, ax in zip(factors, axis):
    image = dlm(img,factors=(factor, factor, 1)).astype(int)
    ax.imshow(image)
    ax.set_title('$N={}$'.format(image.shape[0]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from skimage.transform import downscale_local_mean as dlm
import skimage.io as io
img = io.imread("1.png")
print("Shape :",img.shape)
factors = 3**np.arange(1, 5)
figure, axis = plt.subplots(3,3)
ax=axis.ravel()
ax[0].imshow(img,cmap='brg')
ax[0].axis('off')
ax[0].set_title('Orginal Image \n $N1={}, N2={}, N3={}$'.format(img.shape[0],img.shape[1],img.shape[2]))
print(img[0:60,20:23,1])
for i in range(8):
    i1=i+1
    image = dlm(img,factors=(i1,i1,1),cval=0).astype(int)
    if i1==3:
        print(image[0:20,20,1])
    ax[i1].imshow(image,cmap='brg')
    ax[i1].axis('off')
    ax[i1].set_title('$N1={}, N2={}, N3={}$'.format(image.shape[0],image.shape[1],image.shape[2]))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
def circle_image(x, y):
    X, Y = np.meshgrid(x, y)
    return X**2 + Y**2
factors = 2**np.arange(1, 5)
circ_image = circle_image(np.linspace(0, 1, num=4), 
                          np.linspace(0, 1, num=4))
fig, ax = plt.subplots(1, len(factors), figsize=(15, 4))
for i, k in enumerate(factors):
    bins = np.linspace(0, circ_image.max(), k)
    print(bins)
    image = np.digitize(circ_image, bins)
    image = np.vectorize(bins.tolist().__getitem__)(image-1)
    ax[i].imshow(image)
    ax[i].set_title('$k = {}$'.format(k))
```
