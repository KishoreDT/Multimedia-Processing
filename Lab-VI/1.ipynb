```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
from skimage.filters import unsharp_mask as um
img=cv.imread("2.png",0)
m,n=img.shape
k1=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
img_s1=cv.filter2D(src=img, ddepth=-1, kernel=k1)
k2=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
img_s2=cv.filter2D(src=img, ddepth=-1, kernel=k2)
r1=um(img, radius=5, amount=1)
r2=um(img, radius=10, amount=1)
t=["Orginal Image","Sharpenned Image with k1","Sharpenned Image with k2","Unsharp Mask radius=5","Unsharp Mask radius=10"]
im=[img,img_s1,img_s2,r1,r2]
for i in range(5):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
import skimage.color as cl
img=io.imread("img.png")
img_gr=cl.rgb2gray(img)
img_fft=np.fft.fftshift(np.fft.fft2(img_gr))
img_fft1=np.fft.fftshift(np.fft.fft2(img_gr[0:100,0:100]))
img_recons=abs(np.fft.ifft2(img_fft))
t=["Orginal Image","Grayscaled Image","Magnituded Spectrum 1","Magnituded Spectrum 2","Reconstructed Image"]
im=[img,img_gr,np.log(abs(img_fft)),np.log(abs(img_fft1)),img_recons]
for i in range(1):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
for i in range(1,5):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```
