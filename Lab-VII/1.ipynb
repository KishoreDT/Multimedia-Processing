```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io

img1 = io.imread("1.png")
mask1 = cv.imread("mask1.png", 0)
img2 = io.imread("2.png")
mask2 = cv.imread("mask2.png", 0)
res1 = cv.inpaint(img1, mask1, 3, cv.INPAINT_TELEA)
res2 = cv.inpaint(img2, mask2, 1, cv.INPAINT_NS)
t=["Damaged Image","Masked Image","Restored Image","Image with logo","Masked Image","Image without logo"]
im=[img1,mask1,res1,img2,mask2,res2]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```
