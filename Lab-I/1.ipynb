```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
img = cv.imread("1.png")
img1 = cv.imread("1.png")
img2 = cv.imread("1.png",flags=cv.IMREAD_GRAYSCALE)
print(f"Shape of Image is : {img.shape}")
print(img[1,2])
n=100*np.ones(150)
print(img1[0:100,56,0])
for i in range(45,65):
    img1[0:150,i,1]=n
    img2[0:150,i]=n
print(img1[0:100,46,1])

img3 = cv.imread("1.png")
x=int(input("Enter a number for x-axis :"))
y=int(input("Enter a number for y-axis :"))
#h=int(input("Enter a height :"))
#l=int(input("Enter a length :"))
rs = cv.resize(img3,(x,y))
cp = img[80:280, 150:330]
t=["Orginal","Image","GrayScale","Resize","Croped"]
im=[img,img1,img2,rs,cp]
c=5
for i in range(c):
    plt.subplot(3,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
img = io.imread("1.png")
org = io.imread("1.png")
print(f"Shape of Image is : {img.shape}")
n=100*np.ones(150)
print(img[0:100,56,0])
for i in range(45,65):
    img[0:150,i,1]=n
print(img[0:100,46,1])
t=["Orginal Image","Changed Image"]
im=[org,img]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
plt.show()
```
